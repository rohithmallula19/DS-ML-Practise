```python
import pandas as pd
import numpy as np
```

```python
url= 'https://raw.githubusercontent.com/krishnaik06/Complete-Feature-Engineering/master/mercedesbenz.csv'
df= pd.read_csv(url,index_col=0,usecols=['X0','X1','X2','X3','X4','X5','X6'])
```

```python
df.head(5)
```

```python
for col in df.columns:
  print(col,':',len(df[col].unique()), 'labels')
```

```python
pd.get_dummies(df,drop_first=True).shape
```

```python
df.X2.value_counts().sort_values(ascending= False).head(5)
```

```python
top_10=[x for x in df.X2.value_counts().sort_values(ascending= False).head(10).index]
top_10
```

```python
for label in top_10:
  df[label]= np.where(df['X2']==label,1,0)
```

```python
df[['X2']+top_10].head(20)
```

```python
def one_hot_encoding_top_x(df, variable, top_x_labels):
    
    for label in top_x_labels:
        df[variable+'_'+label] = np.where(df[variable]==label, 1, 0)
```

```python
def one_hot_encoding_top_x(df,variable,top_x_labels):
  for label in top_x_labels:
    df[variable+'_'+label]=np.where(df[[variable]==label,1,0])
```
